```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
plt.rcParams['figure.figsize'] = (14,8)
```

```python
data = np.load('woofer_numpy_log.npz')
est = np.load('woofer_state_est_log.npz')['arr_0']
```

```python
fig, (ax1, ax2,ax3) = plt.subplots(nrows=3, sharex=True)
p_true = data['state_history'][10,:]
l1 = ax1.plot(p_true)
ax1.set_title('state_history')

print(np.shape(est))

p_est = est[10, :]
l2 = ax2.plot(p_est)
ax2.set_title('state_est')

p_diff = p_est - p_true
l3 = ax3.plot(p_diff)
ax3.set_title('state_err')


fig.set_size_inches(14,12)
plt.show()
```
